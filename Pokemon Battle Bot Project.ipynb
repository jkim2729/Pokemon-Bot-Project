### Project Outline

This exploration is an attempt to compute possible number of initial states of a pokemon match along with possible number of total states. Note that all matches must obey gen 1 OU rules according to Smogon as of December 2024. Additionally we are assuming that every pokemon has maxed out EV's and DV's as choosing not to do this is very rare. There are some small use cases for this in Gen 1 Ubers, but in OU it is rarer so we will omit it as adding it in gives many additional scenarios that will almost never happen and many are nearly identical to the perfect case anyway. For instance a pokemon with a 0 on attack EV's but the pokemon only has special moves. Or any EV's that are close to 252 will also usually not matter. 

## Pokemon

There are 151 total pokemon in Generation 1 with 149 being allowed in Gen 1 Overused. Since we are assuming maxed out ev's and dv's each pokemon only has 1 combination of stats. Moves is the next big thing to consider. A pokemon generally has 20+ possible moves to consider. Each pokemon must have at least 1 move. There are no items or abilities in generation 1 so we can avoid those in computation. Lastly, we must consider which pokemon is the lead one. Orders of the other pokemon are irrelevant and will not be counted as they are exactly the same for all intents and purposes. So are calculations are relatively simple. The only unknown at the moment is how many move combinations each pokemon has so we do that now.

```python
def get_pokemon_moves(pokemon_name_or_id):
    # PokéAPI endpoint for Pokémon data
    url = f'https://pokeapi.co/api/v2/pokemon/{pokemon_name_or_id}'
    
    # Make a GET request to the API
    response = requests.get(url)
    
    # If the request was successful (status code 200)
    if response.status_code == 200:
        data = response.json()
        
        # Extract the list of moves from the response data
        moves = data['moves']

        return moves
    else:
        print(f"Failed to retrieve data for {pokemon_name_or_id}. Status code: {response.status_code}")
```

```python
import math
### Formula to compute stat totals, first formula is used for hp with the second being used for the 4 other stats

def calculate_stats(B, I, E, L, use_second_formula=False):
    # Step 1: Calculate sqrt(E)

    sqrt_E = math.sqrt(E)

    # Step 2: Apply ceil to the square root
    ceil_sqrt_E = math.ceil(sqrt_E)

    # Step 3: Divide the result by 4
    division_result = ceil_sqrt_E / 4

    # Step 4: Apply floor to the division result
    floor_division_result = math.floor(division_result)

    # Step 5: Calculate (B + I) * 2
    sum_BI = (B + I) * 2

    # Step 6: Add floor of division result to the sum of B + I * 2
    intermediate_result = sum_BI + floor_division_result

    # Step 7: Multiply by L
    result_L = intermediate_result * L

    # Step 8: Divide by 100
    division_by_100 = result_L / 100

    # Step 9: Apply floor to the result
    floor_result = math.floor(division_by_100)

    # Step 10: Add the final value
    if use_second_formula:
        # Second formula: Add 5
        final_result = floor_result + 5
    else:
        # First formula: Add L + 10
        final_result = floor_result + L + 10

    return final_result




```

```python
import math
def health(total,percent):
    if percent == 100:
        return (total,)
    elif percent == 99:
        max_num = total - 1
        min_num = math.ceil(0.981*total)
    elif percent == 0:
        return (0,)
    else:
        max_num = math.floor(total*percent/100)
        min_num = math.ceil(((percent-0.999)/100)*total)
    
    return tuple(range(min_num,max_num+1))

```

```python
import re
```

```python
# Function to parse the team data
def parse_team_data(team_data):
    # Split the team into blocks (each block corresponds to a Pokémon)
    blocks = re.split(r"\n\n", team_data.strip())

    # Initialize an empty dictionary to store the Pokémon and their moves
    pokemon_team = {}

    # Iterate over each block and extract the Pokémon name and its moves
    for block in blocks:
        lines = block.split("\n")
        
        # Extract the Pokémon name (the first line)
        pokemon_name = lines[0].strip()

        # Extract moves (lines starting with '-')
        moves = [line.strip('-').strip() for line in lines[2:] if line.startswith('-')]
        
        # Add to dictionary
        pokemon_team[pokemon_name] = moves

    return pokemon_team

# Parse the team
team_dict = parse_team_data(team)

# Print the resulting dictionary
print(team_dict)
```
