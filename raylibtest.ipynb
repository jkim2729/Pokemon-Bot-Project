```python
env = Pokebot_Gen1_Environment(battle_format='gen1ou',team1=team_1,team2=team_2,start_challenging=True)
obs = env.reset()

done = {agent: False for agent in env.agents}
while not all(done.values()):
    # Sample random actions for active agents
    actions = {
        agent: env.action_space(agent).sample() for agent in env.agents if not done[agent]
    }

    try:
        obs, rewards, done, truncations, info = env.step(actions)
        print(f"Step complete. Rewards: {rewards}")
    except Exception as e:
        print("Error during step:", e)
        break
```
