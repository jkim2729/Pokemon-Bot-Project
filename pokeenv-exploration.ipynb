```python
class test_bot(Player):
    def __init__(
        self,
        account_configuration: Optional["AccountConfiguration"] = None,
        *,
        avatar: Optional[str] = None,
        battle_format: str = "gen9randombattle",  # Default battle format
        log_level: Optional[int] = None,
        max_concurrent_battles: int = 2,  # Overriding max_concurrent_battles to 2 for CustomBot
        accept_open_team_sheet: bool = False,
        save_replays: Union[bool, str] = False,
        server_configuration: Optional["ServerConfiguration"] = None,
        start_timer_on_battle_start: bool = False,
        start_listening: bool = True,
        ping_interval: Optional[float] = 20.0,
        ping_timeout: Optional[float] = 20.0,
        team: Optional[Union[str, "Teambuilder"]] = None,
        custom_team_name: str = "EpicBot Team",  # Custom argument for subclass
    ):
        # Calling parent constructor to initialize common variables
        super().__init__(
            account_configuration=account_configuration,
            avatar=avatar,
            battle_format=battle_format,
            log_level=log_level,
            max_concurrent_battles=max_concurrent_battles,  # Custom value
            accept_open_team_sheet=accept_open_team_sheet,
            save_replays=save_replays,
            server_configuration=server_configuration,
            start_timer_on_battle_start=start_timer_on_battle_start,
            start_listening=start_listening,
            ping_interval=ping_interval,
            ping_timeout=ping_timeout,
            team=team,
        )
        self.q_table = {}  # Q-table: stores Q-values for (state, action) pairs
        self.learning_rate = 0.1  # alpha
        self.discount_factor = 0.9  # gamma
        self.exploration_rate = 1.0  # epsilon
        self.exploration_decay = 0.995
        self.min_exploration_rate = 0.01
        


    def get_state(self, battle: AbstractBattle):
        player_hp = battle.active_pokemon.current_hp
        player_status = battle.active_pokemon.status
        player_recharge = battle.active_pokemon.must_recharge
        opponent_hp = battle.opponent_active_pokemon.current_hp
        opponent_status = battle.opponent_active_pokemon.status
        opponent_recharge = battle.opponent_active_pokemon.must_recharge

        return (player_hp, player_status, player_recharge, opponent_hp, opponent_status, opponent_recharge)
        
    def get_reward(self, battle: AbstractBattle) -> float:
        """Returns the reward based on battle outcome: +1 for win, 0 for loss."""
        return 1 if battle.won else 0

    def choose_move(self, battle: AbstractBattle) -> BattleOrder:
        """Choose the bot's move using epsilon-greedy strategy."""
        state = self.get_state(battle)
        
        # Initialize Q-values for the state if it's not already in the Q-table
        if state not in self.q_table:
            self.q_table[state] = np.zeros(len(battle.available_moves))

        # Epsilon-greedy strategy: Explore or exploit
        if random.uniform(0, 1) < self.exploration_rate:
            # Exploration: choose a random move
            move = random.choice(battle.available_moves)
        else:
            # Exploitation: choose the move with the highest Q-value
            action = np.argmax(self.q_table[state])
            move = battle.available_moves[action]

        return move

    def learn(self, state, action, reward, next_state, done):
        """Update Q-values after the battle ends."""
        if done:  # Only update Q-values at the end of the battle
            if state not in self.q_table:
                self.q_table[state] = np.zeros(len(battle.available_moves))

            # Q-learning update rule (no future rewards since we're using a final reward)
            self.q_table[state][action] += self.learning_rate * (reward - self.q_table[state][action])

            # Optionally, decay exploration rate over time
            if self.exploration_rate > self.min_exploration_rate:
                self.exploration_rate *= self.exploration_decay

            print(f"Updated Q-table for state {state}: {self.q_table[state]}")  # Debugging output

    def end_battle(self, battle: AbstractBattle, state, action):
        """Call learn method after battle ends to update Q-table based on reward."""
        reward = self.get_reward(battle)
        next_state = self.get_state(battle)
        
        # Mark battle as done (final outcome)
        done = True
        
        # Update Q-values with the result of the battle
        self.learn(state, action, reward, next_state, done)
```
